```python
#rename all columns to lowercase
df.columns = df.columns.str.lower()
df.columns
```

```python
#change index to linename
df.index.name = 'linename'
df.head()
```

```python
#reset index
df.reset_index(drop =True,inplace=True)
df.head()
```

```python
#create a new column
df['Num_Lines'] = df['linename'].apply(len) #how many lines pass through a station
#sort in descending order
df.sort_values('Num_Lines',ascending=False)

df
```

```python
df.columns
```

```python
#clean data
#clean columnn names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.columns.map(lambda x: x.strip())
df.columns
```

```python
#
df['date'].head()
```

```python
pd.to_datetime(df['date']).head()
```

```python
#creating new column day_of_week
df['day_of_week'] = pd.to_datetime(df['date']).dt.day_name()
df.head()
```

```python
df.columns
```

```python
# Group the data by day of week and plot the sum of the numeric columns
grouped = df.groupby('day_of_week').sum(numeric_only=True)
grouped.plot(kind='bar')
plt.show()

```

```python
#remove index of grouped
grouped.reset_index(inplace=True)
grouped
```

```python
#remove index of grouped
grouped.set_index('day_of_week',inplace=True)
grouped
```

```python
grouped.head()
```

```python

#using dictionary weekend_map
weekend_map = {
    'Saturday': True,
    'Sunday': True,
    'Monday': False,
    'Tuesday': False,
    'Wednesday': False,
    'Thursday': False,
    'Friday': False
}
```

```python
#new column of 'is_weekend' that maps 'day_of_weeek'
grouped['is_weekend'] = grouped['day_of_week'].map(weekend_map)
grouped
```

```python
grouped['is_weekend'] = grouped.index.map(weekend_map)
grouped
```

```python
# Group the data by weekend/weekday and plot the sum of the numeric columns
wkend = grouped.groupby('is_weekend').sum(numeric_only = True)
wkend[['entries', 'exits']].plot(kind='bar')
plt.show()
```
